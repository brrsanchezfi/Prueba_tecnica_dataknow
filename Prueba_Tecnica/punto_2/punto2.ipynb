```python
import pandas as pd
```

```python
path = 'Datos Maestros VF.xlsx'
```

```python
columnas_seleccionadas = ['Nombre visible Agente',
                          'AGENTE (OFEI)',
                          'CENTRAL (dDEC, dSEGDES, dPRU…)',
                          'Tipo de central (Hidro, Termo, Filo, Menor)']
```

```python
df = pd.read_excel(path, usecols=columnas_seleccionadas)
```

```python
df.head(5)
```

```python
condicion_agente = df['Nombre visible Agente'].isin(['EMGESA', 'EMGESA S.A.'])
condicion_tipo_central = df['Tipo de central (Hidro, Termo, Filo, Menor)'].isin(['H', 'T'])

resultados_filtrados = df[condicion_agente & condicion_tipo_central]
```

```python
resultados_filtrados
```

```python
df_central = pd.read_csv('dDEC1204.TXT', header=None, encoding='latin1')
df_central
```

```python
df_central[0].nunique() == len(df_central[0])
```

```python
df_merge = pd.merge(resultados_filtrados,df_central,left_on='CENTRAL (dDEC, dSEGDES, dPRU…)', right_on=0, how='inner')
df_merge
```

```python
ultimas_24_columnas = df_merge.iloc[:, -24:]
df_merge['suma'] = ultimas_24_columnas.sum(axis=1)
df_merge
```

```python
resultados_filtrados_2 = df_merge[df_merge['suma']>0]
resultados_filtrados_2
```

```python
# Guardar en un archivo Excel
resultados_filtrados_2.to_excel('resultados.xlsx', index=False)

# Guardar en un archivo CSV
resultados_filtrados_2.to_csv('resultados.csv', index=False)
```
